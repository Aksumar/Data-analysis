```
import pandas as pd
```

```
df = pd.read_csv("pokemon_data.csv")
df.drop(columns = ['#'], inplace = True)
df.head()
```

```
df.info()
```

```
df.describe()
```

### The first way to sum up column`s values using +

```
df['Total'] = df.HP + df.Attack + df.Defense + df['Sp. Atk']+ df['Sp. Def'] + df.Speed

df.sort_values(by = 'Total', ascending=False, inplace=True)
df.head()
```

**RayquazaMega Rayquaza** & **MewtwoMega Mewtwo X** & **MewtwoMega Mewtwo** Y atr the best! They all have 780 points!

### The second way to sum up column's values using iloc and slices

```
df['Total_iloc'] = df.iloc[:, 3:9].sum(axis = 1)
df.head()
```

### Reordering columns

```
columns = list(df.columns)
columns
```

```
df = df[columns[0:3] + [columns[-2]] + columns[3:-2]]
df.head()
```

## Filtering data

```
df_grass = df[df['Type 1'] == 'Grass']
df_grass
```

```
df_grass['Type 1'].unique()
```

```
df_grass_water = df[(df['Type 1'] == 'Grass') | (df['Type 1'] == 'Water')]
df_grass_water
```

```
df_grass_water['Type 1'].unique()
```

### Regex usage

*re.I* means ignore capitalization

```
import re
```

```
df[df['Type 1'].str.contains('fire|water', flags = re.I,  regex = True)]
```

```
df[df['Name'].str.contains('^pi*', regex = True, flags = re.I)]['Name'].unique().tolist()
```

## Conditional Changes


!!!!! Use of loc is important

```
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
```

```
df[df['Type 1'] == 'Flamer']
```

```
df.loc[df['Type 1'] == 'Flamer', 'HP'] = -10
```

```
df[df['Type 1'] == 'Flamer']
```

```
df.loc[df.HP < 100, ['Type 1', 'Type 2']] = ['Test1', 'Test2']
```

```
df[df.HP < 100]
```

## Aggregate statistics

```
df = df[(df['Type 1'] != 'Test1') & (df['Type 2'] != 'Test2')]
```

```
# check that Type 1 = Test1 was deleted
df['Type 1'].unique()
```

```
df.groupby('Type 1').mean().sort_values('Defense', ascending = False)
```

## Working with large amount of data

```
df_temp = pd.DataFrame(columns = df.columns[4:])

# devide .csv on chuncks

for df_i in pd.read_csv('pokemon_data.csv', chunksize = 10 ):
    result = df_i.groupby('Type 1').mean()

    df_temp = pd.concat([df_temp, result])

df_temp
```
