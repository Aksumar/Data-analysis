```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

# Basic Line Graphs


```
x = [0,1,2,3]
y = [0,2, 4, 6]


plt.plot(x,y, label = '2x', color = 'green', linewidth = 3, marker = 'o', 
         markersize = 10, markerfacecolor = 'black', markeredgecolor = 'red')

plt.title('Very simple graph!', fontdict={'fontsize': 15})

plt.xlabel('X', fontsize = 100)
plt.ylabel('Y', fontsize = 20)

plt.xticks(np.arange(0, 6 + 0.5, 0.5))
plt.yticks(np.arange(0, 6 + 0.5 , 0.5))

# !! важно не забыть, так как без legend() не будет отображаться label, даже если указать его 
plt.legend()

plt.show()
```

```
x = np.arange(0, 10)

#resing of graph
plt.figure(figsize = (5, 10))

#The first graph in two parts
# color, marker and linestyle. shorthand notation
plt.plot(x[:6],x[:6]**2, 'gD-', label = 'y = x^2. The first half')
# продолжение графика в другом стиле - пунктиром
plt.plot(x[5:],x[5:]**2, 'gD--', label = 'y = x^2. The second half ')

#the second graph
plt.plot(x,x * 5, 'ro-', label = 'y = 5x')

plt.title('Two graphs on one plot!', fontdict={'fontsize': 15})

plt.xlabel('X')
plt.ylabel('Y')

plt.xticks(np.arange(0, 10))

# !! важно не забыть, так как без legend() не будет отображаться label, даже если указать его 
plt.legend()

plt.savefig('test_graph.jpg', dpi = 200)

plt.show()
```

#Basic Bar Chart

```
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

bars[0].set_hatch('o')

patterns = ['.', 'o', '|']

for bar_i in bars:
  bar_i.set_hatch(patterns.pop(0))
plt.show()
```

# Real data graph

## Gas prices


```
gas_data_url = 'https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv'
```

```
df_gas = pd.read_csv(gas_data_url)
df_gas.head()
```

```
df_gas.info()
```

```
df_gas.Year.min()
```

```
df_gas.Year.max()
```

```
plt.figure(figsize=(10, 5))
plt.title('Gas prices from 1990 to 2008')


plt.plot(df_gas.Year, df_gas.Australia, '.-', label = 'Australia' )
plt.plot(df_gas.Year, df_gas.Canada,  '.-',label = 'Canada')
plt.plot(df_gas.Year, df_gas['South Korea'],  '.-', label = 'South Korea')
plt.legend()

plt.xticks(df_gas.Year[::2], rotation = 'vertical')

plt.xlabel('year')
plt.ylabel('Price in $')


plt.show()
```

```
plt.figure(figsize=(10, 8))
plt.title('Gas prices from 1990 to 2008')

for country in df_gas.columns[1:]:
  plt.plot(df_gas.Year, df_gas[country], '.-', label = country )

plt.legend()

plt.xticks(df_gas.Year[::2], rotation = 'vertical')

plt.xlabel('year')
plt.ylabel('Price in $')

plt.savefig('Gas prices.png', dpi = 200)
plt.show()  




```

## Fifa data

```
fifa_url = 'https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv'
```

```
df_fifa = pd.read_csv(fifa_url, index_col=0)
df_fifa.head()
```

```
df_fifa.info()
```

### Histograms

```
plt.hist(df_fifa.Overall, color = '#abcdef')

plt.title('Distribution of Player Skills in Fafa')
plt.xticks(np.arange(0, 100, 5))

plt.show()
```

```
#bin use
plt.hist(df_fifa.Overall, bins = np.arange(40, 100, 10))
plt.show()
```

### Pie

```
left = df_fifa[df_fifa['Preferred Foot'] == 'Left']['ID'].count()
right = df_fifa[df_fifa['Preferred Foot'] == 'Right']['ID'].count()


plt.pie([left, right], 
        labels = ['Left foot', 'Right foot'],
        colors = ['#abcdef', '#aaabbb'],
        autopct='%.2f %%')

plt.title('Foot preference of FIFA Players')
plt.show()
```

```
# df_fifa.Weight.str[:-3]

df_fifa.Weight = pd.to_numeric(df_fifa.Weight.str.rstrip('lbs'))
```

```
df_fifa['Weight kg'] = df_fifa.Weight * 0.45359237
```

```
df_fifa.head()
```

```
light = df_fifa[df_fifa['Weight kg'] < 60]['ID'].count()
medium = df_fifa[(df_fifa['Weight kg'] >= 60) & (df_fifa['Weight kg'] < 80)]['ID'].count()
heavy = df_fifa[(df_fifa['Weight kg'] >= 80)]['ID'].count()
```

```
plt.style.use('ggplot')

plt.pie([light, medium, heavy], 
        labels = ['Under  60kg', '60-80 kg', ' 80 kg'],
       autopct='%.2f %%', pctdistance = 0.6)

plt.title('Weight distribution of Fifa players in kg')
plt.show()
```

# Box & Whisker plot

```
df_fifa.Club.unique()
```

```
Bray_Wanderers = df_fifa[df_fifa.Club == 'Bray Wanderers']['Overall']
Limerick_FC = df_fifa[df_fifa.Club == 'Limerick FC']['Overall']
Dynamo_Kyiv = df_fifa[df_fifa.Club == 'Dynamo Kyiv']['Overall']
plt.figure(figsize=(5, 7))

plt.boxplot([Bray_Wanderers, Limerick_FC, Dynamo_Kyiv], labels = ['Bray Wanderers', 'Limerick FC', 'Dynamo Kyiv' ])



plt.title('Fifa team comparision')
plt.ylabel('Overall score')
plt.xlabel('Teams')
plt.show()
```
